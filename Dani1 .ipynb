```python
# Misalnya, kita bisa menggunakan metode IQR untuk mendeteksi outliers
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    return df[~((df[column] < (Q1 - 1.5 * IQR)) | (df[column] > (Q3 + 1.5 * IQR)))]

# Menghapus outliers untuk kolom 'heart_rate' sebagai contoh
df = remove_outliers(df, 'heart_rate')
```

```python


import pandas as pd

# Data MIMIC-IV
data_mimic4 = {
    'stay_id': [38675535, 39999230, 33082251, 32885938, 35292602, 32070690, 35983467, 30669099, 37117569],
    'charttime': ['2182-01-25 07:00:00', '2147-09-01 00:00:00', '2121-10-26 01:00:00', '2127-04-20 15:00:00', '2128-04-07 06:00:00', '2196-04-11 01:00:00', '2138-07-06 15:00:00', '2171-08-31 23:45:00', '2131-03-10 09:00:00'],
    'heart_rate': [119.0, 116.0, 56.0, 117.0, 50.0, 48.0, 35.0, 128.0, 56.0],
    'sbp': [None, 85.0, None, None, 89.0, None, None, 83.0, None],
    'dbp': [None, 45.0, None, None, 49.0, None, None, 47.0, None],
    'mbp': [None, 58.0, None, None, 64.0, None, None, 61.0, None],
    'resp_rate': [21.0, 19.0, 16.0, 32.0, 23.0, 16.0, 20.0, 25.0, 15.0],
    'temperature': [39.3, None, None, None, None, None, None, None, None],
    'spo2': [100.0, 95.0, None, 94.0, 98.0, 100.0, 97.0, 100.0, 99.0],
    'glucose': [None, None, None, None, None, None, None, None, None]
}

# Membuat DataFrame dari data MIMIC-IV
df_mimic4 = pd.DataFrame(data_mimic4)

# Normalisasi: Mengubah format waktu menjadi konsisten
df_mimic4['charttime'] = pd.to_datetime(df_mimic4['charttime'], errors='coerce')

# Agregasi: Menghitung rata-rata untuk setiap pengukuran berdasarkan 'stay_id'
aggregated_data = df_mimic4.groupby('stay_id').agg({
    'heart_rate': 'mean',
    'sbp': 'mean',
    'dbp': 'mean',
    'mbp': 'mean',
    'resp_rate': 'mean',
    'temperature': 'mean',
    'spo2': 'mean',
    'glucose': 'mean'
}).reset_index()

print("Data yang Telah Diagregasi Berdasarkan Stay ID:")
print(aggregated_data)
```

```python
# Pembuatan Atribut Baru: Menghitung perubahan detak jantung antar waktu
df_mimic4['heart_rate_change'] = df_mimic4.groupby('stay_id')['heart_rate'].diff()

# Pembuatan Atribut Baru: Menentukan status kesehatan berdasarkan ambang batas
def determine_health_status(row):
    if row['spo2'] < 90:
        return 'Critical'
    elif row['spo2'] < 95:
        return 'Warning'
    else:
        return 'Normal'

df_mimic4['health_status'] = df_mimic4.apply(determine_health_status, axis=1)

# Menampilkan DataFrame yang telah diubah
print("\nDataFrame MIMIC-IV Setelah Normalisasi, Agregasi, dan Pembuatan Atribut Baru:")
print(df_mimic4)
```
